```python
def make_xor_map(x, y, N):
  colorv = np.zeros(N)
  for i in range(N):
    if x[i] * y[i] < 0:
      colorv[i] = 1
  return colorv

N = 1000
X = np.random.rand(N, 2) * 2 -1
Y = make_xor_map(X[:,0], X[:,1], N)
plt.scatter(X[:,0], X[:,1], c=Y)
plt.show()
```
