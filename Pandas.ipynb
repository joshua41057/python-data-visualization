```python
import matplotlib.pyplot as plt
def circle_pos(r, n, label):
    x0 = np.multiply(4 * np.random.random(n) + r, np.sin(ang))
    y0 = np.multiply(4 * np.random.random(n) + r, np.cos(ang))
    z0 = label * np.ones(N)
    return x0, y0, z0
 
 
N = 1000
ang = np.linspace(0, 2 * np.pi, N)
x_in, y_in, label_in = circle_pos(r=5, n=N, label=0)
x_out, y_out, label_out = circle_pos(r=9, n=N, label=1)
x1 = np.concatenate((x_in, x_out), axis=0)
x2 = np.concatenate((y_in, y_out), axis=0)
y = np.concatenate((label_in, label_out), axis=0)
plt.figure(figsize=(6, 6))
plt.scatter(x1, x2, c=y)
plt.xlabel("x1")
plt.ylabel("x2")
plt.show()
 
data = {
    "x1": x1,
    "x2": x2,
    "y": y
}
 
df = pd.DataFrame(data)
 
def pow2_x1(row):
    return row["x1"] * row["x1"]
 
def pow2_x2(row):
    return row["x2"] * row["x2"]
 
def product(row):
    return row["x1"] * row["x2"]
 
df["x1^2"] = df.apply(pow2_x1, axis=1)
df["x2^2"] = df.apply(pow2_x2, axis=1)
df["x1*x2"] = df.apply(product, axis=1)
 
df.to_csv("./output.csv", header=False, index=False)
```
